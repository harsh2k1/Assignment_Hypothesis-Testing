# Task 1 - Use decision trees to prepare a model on fraud data treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.metrics import classification_report , f1_score , confusion_matrix , accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('data/Fraud_check.csv')
df.head()
```

```python
label = LabelEncoder()
```

```python
df['Undergrad'] = label.fit_transform(df['Undergrad'])
df['Urban'] = label.fit_transform(df['Urban'])
df = pd.get_dummies(df, columns = ['Marital.Status'], drop_first= True)
```

```python
df.head()
```

```python
df['Taxable.Income']
```

```python
#Creating new cols TaxInc and dividing 'Taxable.Income' cols on the basis of [10002,30000,99620] for Risky and Good
df["TaxInc"] = pd.cut(df["Taxable.Income"], bins = [10002,30000,99620], labels = ["Risky", "Good"])
```

```python
df[df['Taxable.Income'] >= 30000]
```

```python
df[df['Taxable.Income'] < 30000]
```

```python
df['TaxInc'] = label.fit_transform(df['TaxInc'])
```

```python
plt.figure(figsize = (13,5))
sns.heatmap(df.corr() , cmap = 'Blues' , annot = True)
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(df.iloc[:,:])
df_norm.tail(10)
```

```python
# Separating target column
x = df_norm.iloc[: , :-1]
y = df_norm.iloc[:, -1]
```

```python
# Create parameter grid  
param_grid = {
    'max_depth': [60, 90, 110],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300]
}

# Instantiate the model
clf_rf = RandomForestClassifier()

# Instantiate grid search model
grid_search = GridSearchCV(estimator = clf_rf, param_grid = param_grid)

grid_search.fit(x,y)
grid_search.best_params_
```

```python
kfold = KFold(n_splits=10, random_state=7 , shuffle = True)
model = RandomForestClassifier(max_depth= 60 , min_samples_leaf= 3 , min_samples_split= 10 , n_estimators = 200)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```

```python
model.fit(x,y)
```

## Plotting 5th estimator (Decision Tree) out of total 200

```python
import graphviz
from sklearn.tree import export_graphviz

dot_data = export_graphviz(model.estimators_[4], feature_names = ['Undergrad', 'Taxable.Income', 'City.Population', \
                                                                  'Work.Experience','Urban', 'Marital.Status_Married', \
                                                                  'Marital.Status_Single'],class_names = ['Risky','Good'], \
                           filled=True, impurity=True, rounded=True)

graph = graphviz.Source(dot_data, format='png')
graph
```

### It is concluded that the Random forest model with {'max_depth': 60, 'min_samples_leaf': 3, 'min_samples_split': 10, 'n_estimators': 200} is best fit.<br> Mean Cross validation score for this model is 0.998
