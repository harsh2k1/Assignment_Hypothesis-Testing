# Task - 2 : A cloth manufacturing company is interested to know about the segment or attributes causes high sale. Approach - A decision tree can be built with target variable Sale (we will first convert it in categorical variable) & all other variable will be independent in the analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report , f1_score , confusion_matrix , accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('data/Companydata.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
label = LabelEncoder()
```

```python
# Label Encoding
df['Urban'] = label.fit_transform(df['Urban'])
df['US'] = label.fit_transform(df['US'])
df = pd.get_dummies(df , columns= ['ShelveLoc'] , drop_first= True)
df.head()
```

```python
df['Sales'].mean()
```

```python
# Converting sales to categorical variable
df['Sales'] = pd.cut(df["Sales"], bins = [df['Sales'].min() , df['Sales'].mean() , df['Sales'].max()], labels = ["Bad", "Good"])
```

```python
df.head()
```

```python
df['Sales'] = label.fit_transform(df['Sales'])
df.head()
```

```python
plt.figure(figsize = (13,5))
sns.heatmap(df.corr() , cmap = 'Blues' , annot = True)
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(df.iloc[:,1:])
df_norm.tail(10)
```

```python
# Separating target column
x = df_norm.iloc[: , :]
y = df.iloc[:,0]
```

```python
# Splitting into training and testing data
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size= 0.20 , random_state= 42)
print(x_train.shape , '\n', x_test.shape ,'\n' , y_train.shape , '\n' ,y_test.shape)
```

## Finding best parameters for Decision tree model using Grid Search

```python
model = DecisionTreeClassifier()
criterion = ['gini' , 'entropy']
max_depth = [4,6,8,10,12]
parameters = {'criterion': criterion , 'max_depth': max_depth}
clf_GS = GridSearchCV(estimator = model, param_grid=parameters)
clf_GS.fit(x_train, y_train)
```

```python
clf_GS.best_params_
```

## Decision tree model using criteria =entropy and max depth = 4

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=4)
model.fit(x_train,y_train)
```

```python
fn= df.columns
cn=['Good', 'Bad']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (6,6), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
# Predicting on training data
y_pred = model.predict(x_train)
print('Predicted value counts: \n',pd.Series(y_pred).value_counts())
print('-'*50)
print('Crosstab: \n',pd.crosstab(y_train,y_pred))
print('-'*50)
print('Classification Report: \n', classification_report(y_train , y_pred))
print('-'*50)
print('Confusion Matrix: \n', confusion_matrix(y_train , y_pred))
print('-'*50)      
print('Accuracy Score = ',accuracy_score(y_train , y_pred))
```

```python
# Predicting on testing data
y_pred = model.predict(x_test)
print('Predicted value counts: \n',pd.Series(y_pred).value_counts())
print('-'*50)
print('Crosstab: \n',pd.crosstab(y_test,y_pred))
print('-'*50)
print('Classification Report: \n', classification_report(y_test , y_pred))
print('-'*50)
print('Confusion Matrix: \n', confusion_matrix(y_test , y_pred))
print('-'*50) 
print('Accuracy Score = ',accuracy_score(y_test , y_pred))
```

### Hyperparameter tuning is a great way to determine the best params for the decision tree based models. I used GridSearchCV method for this purpose. Training and Testing accuracy are comparable. Hence , the model performed well with the parameters I fed to it.
